```python
import pandas as pd

data = pd.read_csv("https://raw.githubusercontent.com/dragonnomada/pycien-2023/main/datasets/weight-height.csv")

data.sample(5)
```

```python
import seaborn as sns

sns.scatterplot(x=data["Height"], y=data["Weight"], hue=data["Gender"])
```

```python
def dataviz_2x_1y(x1, x2, y, **options):
  import matplotlib.pyplot as plt
  import seaborn as sns

  sns.scatterplot(x=x1, y=x2, hue=y)

  if "title" in options:
    plt.title(options["title"])

  if "xlabel" in options:
    plt.xlabel(options["xlabel"])

  if "ylabel" in options:
    plt.ylabel(options["ylabel"])

  if "filename" in options:
    plt.savefig(options["filename"])

  plt.show()
```

```python
dataviz_2x_1y(data["Height"], data["Weight"], data["Gender"], title="Altura vs Peso", xlabel="Altura", ylabel="Peso", filename="fig1.png")
```

```python
data2 = pd.read_csv("https://github.com/dragonnomada/pycien-2023/raw/main/datasets/california_housing_train.csv")

data2.sample(5)
```

```python
dataviz_2x_1y(data2["total_rooms"], data2["population"], data2["households"])
```
