````python
##
#
# Calcular los primeros 20 n칰meros primos
#
# Por:      Alan Badillo Salas
# Contacto: alan@nomadacode.com
# Creado:   Martes 6, junio 2023
#
##

primos = [2] # Guardamos los n칰meros primos encontrados en una lista

n = 2 # Creamos una variable que almacene el pr칩ximo n칰mero a verificar
      # si es un n칰mero primo (partimos del 2)

while len(primos) < 20: # Repetimos mientras la lista no tenga 20 elementos

  n += 1 # Incrementamos el siguiente n칰mero a determinar si es primo

  esPrimo = True # Creamos una variable que suponga que el n칰mero es primo

  for p in primos: # Recorremos cada primo ya calculado de la lista
    
    if n % p == 0: # Determinamos si el n칰mero es m칰ltiplo de alg칰n primo
      esPrimo = False # Si el n칰mero es m칰ltiplo de alg칰n primo,
                      # ya no puede ser un n칰mero primo, la suposici칩n era falsa
  
  if esPrimo: # Determinamos si la variable mantiene True,
              # esto significa que el n칰mero no es m칰ltiplo de otro primo,
              # por lo tanto es un n칰mero primo y lo guardamos
    primos.append(n)

print(primos) # Imprimimos la lista de los n칰meros primos encontrados
              # [2, 3, 5, 7, 11, 13, 17, 19, ...]
````

### Ejemplo de operaciones aritm칠ticas para calcular la distancia entre dos puntos

La distancia entre dos puntos se puede obtener mediante el Teorema de Pit치goras, formando con los dos puntos una recta que represente la hipotenusa de un tri치ngulo rect치ngulocomo en la siguiente imagen:

````python
# Establecemos el valor de los puntos (x1, y1) y (x2, y2)

x1 = 2
y1 = 3

x2 = 5
y2 = 9

# Calculamos la diferencia de las x-es y las y-es

dx = x2 - x1
dy = y2 - y1

# Aplicamos el teorema de pit치goras

d = (dx ** 2 + dy ** 2) ** 0.5 # Observa que la ra칤z cuadrada equivale a elevar
                               # todo a 1/2

print("Distancia entre ({:.2f}, {:.2f}) y ({:.2f}, {:.2f}) = {:.2f}".format(x1, y1, x2, y2, d))
````

````python
import math

f = lambda x: 3 * math.cos(x) - 2

a = 0
b = 5

n = 1000

delta = (b - a) / (n - 1)

suma = 0

for i in range(0, n):
  xi = a + delta * i
  xi_1 = a + delta * (i + 1)

  hi = f(xi)
  hi_1 = f(xi_1)

  base = xi_1 - xi # delta
  altura = (hi + hi_1) / 2

  area = base * altura

  suma = suma + area

print(suma)
````

````python
def distancia(x1, y1, x2, y2):
  dx = x2 - x1
  dy = y2 - y1
  d = (dx ** 2 + dy ** 2) ** 0.5
  return d

d1 = distancia(0, 0, 1, 1)
d2 = distancia(2, 3, 5, 7)
d3 = distancia(-1, 4, 25, -100)

d1, d2, d3
````

### Ejemplo de una funci칩n que imprime el reporte de una lista de valores

````python
def reporte_estadistico(lista):
  # Suma
  s = 0
  
  for x in lista:
    s += x

  # Total
  n = len(lista)

  # Promedio
  p = s / n

  # M칤nimo
  mini = lista[0]

  for x in lista:
    if x < mini:
      mini = x

  # M치ximo
  maxi = lista[0]

  for x in lista:
    if x > maxi:
      maxi = x

  print("LISTA: {}".format(lista))
  print("----------------------")
  print("TOTAL: {}".format(n))
  print("----------------------")
  print("SUMA: {}".format(s))
  print("----------------------")
  print("M칈N: {} | M츼X: {}".format(mini, maxi))
  print("----------------------")
  print("PROMEDIO: {:.2f}".format(p))
  print("----------------------")

print("REPORTE 1:\n")

reporte_estadistico([1, 2, 3, 4, 5, 6])

print("\nREPORTE 2:\n")

reporte_estadistico([3, 2, 5, 4, 9, 1, 7, 6])
````

## Mapeos y filtros

Los mapeos y filtros son operaciones que podemos aplicar a las listas para transformar o reducir los elementos de manera sistem치tica.

Por ejemplo:

* Dada una lista de n칰meros podr칤amos elevar cada uno al cuadrado (transformaci칩n o mapeo)
* Dada una lista de n칰meros podr칤amos dejar s칩lo aquellos que sean pares (reducci칩n o filtrado)
* Dada una lista de n칰meros podr칤amos restar a cada uno una constante (transformaci칩n o mapeo)
* Dada una lista de n칰meros podr칤amos dejar s칩lo aquellos mayores a una constante (reducci칩n o filtrado)
* Dada una lista de textos podr칤amos convertir cada uno a su n칰mero de caracteres (transformaci칩n o mapeo)
* Dada una lista de textos podr칤amos dejar aquellos que contienen n칰meros (reducci칩n o filtrado)

Los mapeos se pueden conseguir usando una funci칩n de transformaci칩n ($T$) y una lista a transformar ($x$) obteniendo una nueva lista ($y$).

La ecuaci칩n ser칤a: $y = map(T, x)$

Donde:

$T: x_i  y_i$

Los filtros se pueden conseguir usando una funci칩n de filtrado ($F$) y una lista a transformar ($x$) obteniendo una lista reducida ($x'$).

La ecuaci칩n ser칤a: $x' = filter(F, x)$

Donde:

$F: x_i  true | false$

### Ejemplo del mapeo de elementos al cuadrado

````python
x = [1, 2, 3, 4, 5, 6]

T = lambda xi: xi ** 2

y = list( map(T, x) ) # Observa que reconvertimos el mapeo a una lista

print(x)
print(y)
````

### Ejemplo del filtrado de elementos pares

````python
x = [1, 2, 3, 4, 5, 6]

F = lambda xi: xi % 2 == 0

x_prima = list( filter(F, x) ) # Observa que reconvertimos el mapeo a una lista

print(x)
print(x_prima)
````

### Ejemplo de mapeos para obtener la varianza de una lista de n칰meros

La varianza es: 

$VAR(x) = \frac{1}{n} 갬_{x_i 갣 x} (x_i - \overline{x})^2$

La desviaci칩n est치ndar es:

$STD.DEV(x) = \sqrt{VAR(x)}$

````python
tornillos = [1.01, 1.03, 0.98, 0.99, 1.0, 1.005, 1.03, 0.98]

promedio = sum(tornillos) / len(tornillos)

T = lambda ti: (ti - promedio) ** 2

varianza = sum(map(T, tornillos)) / len(tornillos)

varianza
````

### Reporte Impreso 1 - Tabla de los Senos y Cosenos

````python
# Desde la librer칤a math importamos el valor de pi
from math import pi

# Llenaremos una lista de puntos `xi`
x = []

# Los puntos estar치n entre `a` y `b`

# L칤mite inferior
a = -pi
# L칤mite superior
b = pi

# Total de puntos
n = 20

# Recorremos los `n` puntos
for i in range(0, n):
  d = (b - a) / n # Factor delta 
                  # (diferencia del intervalo entre el n칰mero de puntos)
  
  xi = a + i * d # Calculamos el punto `xi`

  x.append(xi) # Lo agregamos a la lista

print(x) # Imprimimos los `n = 20` puntos generados entre (`a = -`, `b = `)
````

````python
# Desde la librer칤a math importamos la funci칩n seno y coseno
from math import sin, cos

y1 = list(map(sin, x))

y2 = list(map(cos, x))

print(y1)
print(y2)
````

En este reporte usaremos la lista con los nombres de productos, la lista de sus precios y la lista de sus existencias, para as칤 generar un reporte de productos.

````python
# Lista de nombres
nombres = ["Coca Cola", "Galletas Mar칤as", "Gansito", "Pepsi"]

# Lista de precios
precios = [54.99, 17.81, 21.78, 45.89]

# Lista de existencias
existencias = [2450, 273, 1501, 998]

# Abrimos el archivo para escribir el reporte
f = open("reporte_productos.txt", "w")

# Unimos las tres listas como una sola (de tuplas)
productos = zip(nombres, precios, existencias)

# Calculamos el separador
f.write("+{}+{}+{}+".format("-" * 22, "-" * 10, "-" * 13))
f.write("\n")
f.write("| {:20} | {:8} | {:10} |".format("NOMBRE", "PRECIO", "EXISTENCIAS"))
f.write("\n")
f.write("+{}+{}+{}+".format("-" * 22, "-" * 10, "-" * 13))
f.write("\n")

for nombre, precio, existencias in productos:
  f.write("| {:20} | ${:>7.2f} | {:>11} |".format(nombre, precio, existencias))
  f.write("\n")

f.write("+{}+{}+{}+".format("-" * 22, "-" * 10, "-" * 13))

f.close()
````
