```python
import scrapy
import scrapydo

scrapydo.setup()

def procesar_html(response):
  print("Página cargada: %s" % response.url)

  # nombre = response.xpath("//h3[starts-with(@class, 'CardProduct_h4')]/text()").getall()
  # precio = response.xpath("//p[starts-with(@class, 'CardProduct_precio1')]/text()").getall()

  # print(nombre)
  # print(precio)

  for nodoPadre in response.xpath("//article[starts-with(@class, 'CardProduct_cardProduct')]"):
    nombre = nodoPadre.xpath(".//h3[starts-with(@class, 'CardProduct_h4')]/text()").get()
    precio = nodoPadre.xpath(".//p[starts-with(@class, 'CardProduct_precio')]/text()").get()
    estrellas = nodoPadre.xpath(".//p[starts-with(@class, 'StarFive_comments')]/text()").getall()
    if estrellas == None:
      estrellas = 0
    else:
      estrellas = "".join(estrellas)
    print(nombre)
    print(precio)
    print(estrellas)
    print("-" * 70)

    yield {
        "nombre": nombre,
        "precio": precio,
        "estrellas": estrellas,
    }

result = scrapydo.crawl("https://www.sanborns.com.mx", callback=procesar_html)

print(result)

import pandas as pd

data = pd.DataFrame(result)

data.sample(10)
```
