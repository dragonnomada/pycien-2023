## Creación de Algoritmos

Dada una función que representa el área de un terreno, buscar minimizar la longitud del terreno dada la siguiente función:

$f(x) = (x - 5) \cdot (x - 3) + 100$

### 1. Visualizar la función

```python
import numpy as np

x = np.linspace(-10, 10, 20)

x
```

```python
y = (x - 5) * (x - 3) + 100

y
```

```python
import matplotlib.pyplot as plt

plt.plot(x, y)
```

```python
f = lambda x: (x - 5) * (x - 3) + 100

h = 0.00001

fp = lambda x: (f(x + h) - f(x)) / h

fpp = lambda x: (fp(x + h) - fp(x)) / h
```

```python
xi = -100

for i in range(1, 6):
  if abs(fpp(xi)) > 0.0000000000000001:
    xj = xi - fp(xi) / fpp(xi)
    print("{} -> {}".format(xi, xj))
    xi = xj
```

```python
import math

f = lambda x: -3 * math.exp(0.1 * x) + x * math.cos(x)

h = 0.00001

fp = lambda x: (f(x + h) - f(x)) / h

fpp = lambda x: (fp(x + h) - fp(x)) / h

xi = 0

for i in range(1, 10):
  if abs(fpp(xi)) > 0.0000000000000001:
    xj = xi - fp(xi) / fpp(xi)
    print("{} -> {}".format(xi, xj))
    xi = xj
```
