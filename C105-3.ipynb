```python
import scrapydo

scrapydo.setup()
```

```python
def parser(response):
  th_budget = response.xpath("//th[starts-with(text(), 'Budget')]")
  print(th_budget)

  for tr in th_budget.xpath("../../tr"):
    # print(tr)
    for td in tr.xpath("./td"):
      data = td.xpath("./text()").get()
      print(data)
```

```python
scrapydo.crawl("https://en.wikipedia.org/wiki/List_of_Star_Wars_films", callback=parser)
```
