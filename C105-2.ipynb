```python
import pandas as pd

data = pd.read_csv("https://raw.githubusercontent.com/dragonnomada/pycien-2023/main/datasets/sanborns_2023_06_30.csv")

data.sample(5)
```

```python
def sanborns_clean(data):
  def fix_columns(columns):
    data2 = data[ columns.keys() ]
    data2.columns = columns.values()
    return data2
  data = fix_columns({ "title": "CATEGORIA", "name": "NOMBRE", "stars": "ESTRELLAS", "price": "PRECIO" })
  def fix_column_nan_init(name):
    data2 = data.copy()
    data2[name] = data[name].fillna(0).map(int)
    return data2
  data = fix_column_nan_init("ESTRELLAS")
  def fix_column_currency(name):
    data2 = data.copy()
    data2[name] = data[name].map(lambda price: price.replace("$", "").replace(",", "")).map(float)
    return data2
  data = fix_column_currency("PRECIO")
  return data
```

```python
sanborns_clean(data)
```
