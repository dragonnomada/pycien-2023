### Sintaxis de Python

Veamos un ejemplo de sintaxis en Python.

> El siguiente ejemplo muestra un programa en Python que calcula los primeros `20` n√∫meros primos.

````python
##
#
# Calcular los primeros 20 n√∫meros primos
#
# Por:      Alan Badillo Salas
# Contacto: alan@nomadacode.com
# Creado:   Martes 6, junio 2023
#
##

primos = [2] # Guardamos los n√∫meros primos encontrados en una lista

n = 2 # Creamos una variable que almacene el pr√≥ximo n√∫mero a verificar
      # si es un n√∫mero primo (partimos del 2)

while len(primos) < 20: # Repetimos mientras la lista no tenga 20 elementos

  n += 1 # Incrementamos el siguiente n√∫mero a determinar si es primo

  esPrimo = True # Creamos una variable que suponga que el n√∫mero es primo

  for p in primos: # Recorremos cada primo ya calculado de la lista
    
    if n % p == 0: # Determinamos si el n√∫mero es m√∫ltiplo de alg√∫n primo
      esPrimo = False # Si el n√∫mero es m√∫ltiplo de alg√∫n primo,
                      # ya no puede ser un n√∫mero primo, la suposici√≥n era falsa
  
  if esPrimo: # Determinamos si la variable mantiene True,
              # esto significa que el n√∫mero no es m√∫ltiplo de otro primo,
              # por lo tanto es un n√∫mero primo y lo guardamos
    primos.append(n)

print(primos) # Imprimimos la lista de los n√∫meros primos encontrados
              # [2, 3, 5, 7, 11, 13, 17, 19, ...]
````

En el c√≥digo anterior podemos observar: 

* Que hay comentarios que ser√°n ignorados despu√©s del `#`.
* La sintaxis de asignaci√≥n de variables usando el s√≠mbolo de igual (`variable = valor`).
* El uso de listas de valores que pueden seguir creciendo como `primos = [2]` y `primos.append(n)`.
* La repetici√≥n de un bloque de c√≥digo mediante `while`.
* La funci√≥n `len(primos)` que devuelve el n√∫mero de elementos en la lista.
* La identaci√≥n despu√©s de los dos puntos (`:`) obligatoria, para marcar pertenencia al bloque de c√≥digo del `while`.
* El incremento en una unidad de la variable `n` (`n += 1`).
* La repetici√≥n de un bloque de sentencias mediante `for iterador in secuencia:`.
* Determinar si ejecutar o no un bloque de c√≥digo mediante `if condici√≥n:`.
* El uso de una variable que contiene `True` o `False`.
* La evaluaci√≥n de un bloque condicionado al valor de `True` o `False` mediante `if valor l√≥gico:`.
* La anidaci√≥n de sentencias usando el tabulador, para indicar si la sentencia pertenece al bloque del `while`, del `for` o de alg√∫n `if`. Marcando el inicio del bloque por los dos puntos `estructura de control:`.

Estas pueden ser utilizadas para generar mensajes que integren datos, extendiendo los textos a formatos.

En Python es indistinto utilizar comillas dobles o simples y se permiten ambas formas para alternar. Es decir, las comillas dobles ignorar√°n a las comillas simples contenidas y viceversa.

### Ejemplo de operaciones aritm√©ticas para calcular la distancia entre dos puntos

La distancia entre dos puntos se puede obtener mediante el Teorema de Pit√°goras, formando con los dos puntos una recta que represente la hipotenusa de un tri√°ngulo rect√°ngulocomo en la siguiente imagen:

![Distancia entre dos puntos a trav√©s de un tri√°ngulo](https://www.geometriaanalitica.info/wp-content/uploads/2020/09/geometria-formula-distancia-entre-dos-puntos.png)

````python
# Establecemos el valor de los puntos (x1, y1) y (x2, y2)

x1 = 2
y1 = 3

x2 = 5
y2 = 9

# Calculamos la diferencia de las x-es y las y-es

dx = x2 - x1
dy = y2 - y1

# Aplicamos el teorema de pit√°goras

d = (dx ** 2 + dy ** 2) ** 0.5 # Observa que la ra√≠z cuadrada equivale a elevar
                               # todo a 1/2

print("Distancia entre ({:.2f}, {:.2f}) y ({:.2f}, {:.2f}) = {:.2f}".format(x1, y1, x2, y2, d))
````

## Condicionales y expresiones l√≥gicas

La condicional es una estructura de control que basado en un valor l√≥gico (o condici√≥n) determina si un bloque de c√≥digo ser√° ejecutado.

El bloque de c√≥digo es un conjunto de sentencias anidadas mediante indentaci√≥n (espacios o tabuladores uniformes que desplazan las l√≠neas de c√≥digo a la derecha).

La sintaxis de un bloque de c√≥digo es:

```txt
ESTRUCTURA:
  BLOQUE

ESTRUCTURA:
  SENTENCIA 1
  SENTENCIA 2
  SENTENCIA 3
  ...

ESTRUCTURA:
  SENTENCIA DENTRO DE LA ESTRUCTURA

SENTENCIA FUERA DE LA ESTRUCTURA
```

Observa que es sumamente importante que la indentaci√≥n sea uniforme durante todo el c√≥digo, pueden ser dos espacios, cuatro espacios, un tabulador, etc.

La condicional se representa como:

> Sintaxis de la condicional simple

```txt
if <condici√≥n o valor l√≥gico>:
  <bloque>
```

A esta se le conoce como condicional simple y hay m√°s variantes.

Los operadores para realizar operaciones l√≥gicas entre valores son:

| Operador | Nombre | Descripci√≥n |
| :------: | ------ | ----------- |
| `>`| Mayor estricto | Determina si un valor es superior a otro $a \gt b$ |
| `>=`| Mayor o igual | Determina si un valor es superior o igual a otro $a \geq b$ |
| `<`| Menor estricto | Determina si un valor es inferior a otro $a \lt b$ |
| `<`| Menor o igual | Determina si un valor es inferior o igual a otro $a \leq b$ |
| `==`| Igual | Determina si un valor es igual a otro $a = b$ |
| `!=`| Distinto | Determina si un valor no es igual a otro $a \neq b$ |
| `and`| Y (conjunci√≥n) | Determina si la primer condici√≥n se cumple y la segunta tambi√©n $p \land q$ |
| `or`| O (disyunci√≥n) | Determina si la primer condici√≥n se cumple o la segunta lo hace $p \lor q$ |

### Ejemplo de una condicional simple

### Ejemplo de una funci√≥n que calcula la distancia entre dos puntos

````python
def distancia(x1, y1, x2, y2):
  dx = x2 - x1
  dy = y2 - y1
  d = (dx ** 2 + dy ** 2) ** 0.5
  return d

d1 = distancia(0, 0, 1, 1)
d2 = distancia(2, 3, 5, 7)
d3 = distancia(-1, 4, 25, -100)

d1, d2, d3
````

### Ejemplo de una funci√≥n que imprime el reporte de una lista de valores

````python
def reporte_estadistico(lista):
  # Suma
  s = 0
  
  for x in lista:
    s += x

  # Total
  n = len(lista)

  # Promedio
  p = s / n

  # M√≠nimo
  mini = lista[0]

  for x in lista:
    if x < mini:
      mini = x

  # M√°ximo
  maxi = lista[0]

  for x in lista:
    if x > maxi:
      maxi = x

  print("LISTA: {}".format(lista))
  print("----------------------")
  print("TOTAL: {}".format(n))
  print("----------------------")
  print("SUMA: {}".format(s))
  print("----------------------")
  print("M√çN: {} | M√ÅX: {}".format(mini, maxi))
  print("----------------------")
  print("PROMEDIO: {:.2f}".format(p))
  print("----------------------")

print("REPORTE 1:\n")

reporte_estadistico([1, 2, 3, 4, 5, 6])

print("\nREPORTE 2:\n")

reporte_estadistico([3, 2, 5, 4, 9, 1, 7, 6])
````

## Mapeos y filtros

Los mapeos y filtros son operaciones que podemos aplicar a las listas para transformar o reducir los elementos de manera sistem√°tica.

Por ejemplo:

* Dada una lista de n√∫meros podr√≠amos elevar cada uno al cuadrado (transformaci√≥n o mapeo)
* Dada una lista de n√∫meros podr√≠amos dejar s√≥lo aquellos que sean pares (reducci√≥n o filtrado)
* Dada una lista de n√∫meros podr√≠amos restar a cada uno una constante (transformaci√≥n o mapeo)
* Dada una lista de n√∫meros podr√≠amos dejar s√≥lo aquellos mayores a una constante (reducci√≥n o filtrado)
* Dada una lista de textos podr√≠amos convertir cada uno a su n√∫mero de caracteres (transformaci√≥n o mapeo)
* Dada una lista de textos podr√≠amos dejar aquellos que contienen n√∫meros (reducci√≥n o filtrado)

Los mapeos se pueden conseguir usando una funci√≥n de transformaci√≥n ($T$) y una lista a transformar ($x$) obteniendo una nueva lista ($y$).

La ecuaci√≥n ser√≠a: $y = map(T, x)$

Donde:

$T: x_i ‚Üí y_i$

Los filtros se pueden conseguir usando una funci√≥n de filtrado ($F$) y una lista a transformar ($x$) obteniendo una lista reducida ($x'$).

La ecuaci√≥n ser√≠a: $x' = filter(F, x)$

Donde:

$F: x_i ‚Üí true | false$

### Ejemplo del mapeo de elementos al cuadrado

````python
x = [1, 2, 3, 4, 5, 6]

T = lambda xi: xi ** 2

y = list( map(T, x) ) # Observa que reconvertimos el mapeo a una lista

print(x)
print(y)
````

### Ejemplo del filtrado de elementos pares

````python
x = [1, 2, 3, 4, 5, 6]

F = lambda xi: xi % 2 == 0

x_prima = list( filter(F, x) ) # Observa que reconvertimos el mapeo a una lista

print(x)
print(x_prima)
````

### Ejemplo de mapeos para obtener la varianza de una lista de n√∫meros

La varianza es: 

$VAR(x) = \frac{1}{n} ‚àë_{x_i ‚àà x} (x_i - \overline{x})^2$

La desviaci√≥n est√°ndar es:

$STD.DEV(x) = \sqrt{VAR(x)}$

````python
tornillos = [1.01, 1.03, 0.98, 0.99, 1.0, 1.005, 1.03, 0.98]

promedio = sum(tornillos) / len(tornillos)

T = lambda ti: (ti - promedio) ** 2

varianza = sum(map(T, tornillos)) / len(tornillos)

varianza
````

````python
# El mapeo representa cada medida de tornillo menos el promedio 
# elevados al cuadrado
list(map(T, tornillos))
````

## Lectura y escritura de archivos

Los archivos son cadenas enormes de caracteres, es decir, textos, o secuencias enormes de bits, es decir bytes.

Podemos extraer el contenido de un archivo de texto mediante la funci√≥n `open(<ruta>, <modo>)`.

| Modo | M√©todos | Descripci√≥n |

## Generaci√≥n de reportes impresos

### Reporte Impreso 1 - Tabla de los Senos y Cosenos

````python
# Desde la librer√≠a math importamos el valor de pi
from math import pi

# Llenaremos una lista de puntos `xi`
x = []

# Los puntos estar√°n entre `a` y `b`

# L√≠mite inferior
a = -pi
# L√≠mite superior
b = pi

# Total de puntos
n = 20

# Recorremos los `n` puntos
for i in range(0, n):
  d = (b - a) / n # Factor delta 
                  # (diferencia del intervalo entre el n√∫mero de puntos)
  
  xi = a + i * d # Calculamos el punto `xi`

  x.append(xi) # Lo agregamos a la lista

print(x) # Imprimimos los `n = 20` puntos generados entre (`a = -œÄ`, `b = œÄ`)
````

````python
# Desde la librer√≠a math importamos la funci√≥n seno y coseno
from math import sin, cos

y1 = list(map(sin, x))

y2 = list(map(cos, x))

print(y1)
print(y2)
````

````python
# Recorremos los √≠ndices del 0 al `n - 1`

print("+--------------------------------+")
print("| TABLA DE SENOS Y COSENOS       |")
print("+----------+----------+----------+")
print("|    x     |  sin(x)  |  cos(x)  |")
print("+----------+----------+----------+")

for i in range(0, n):
  print("| {:^8.2f} | {:^8.2f} | {:^8.2f} |".format(x[i], y1[i], y2[i]))

print("+----------+----------+----------+")
````

## Generaci√≥n de reportes de texto

````python
# Definimos la funci√≥n log

def log(text):
  from datetime import datetime
  message = "LOG {}: {}".format(datetime.now(), text)
  print(message)
  file_log = open("log.txt", "a")
  file_log.write(message)
  file_log.write("\n")
  file_log.close()

log("Hello world")
log("Hola mundo mundial")
log("Oki doki")
````

### Reporte de Texto 1 - Informaci√≥n de productos

En este reporte usaremos la lista con los nombres de productos, la lista de sus precios y la lista de sus existencias, para as√≠ generar un reporte de productos.

````python
# Lista de nombres
nombres = ["Coca Cola", "Galletas Mar√≠as", "Gansito", "Pepsi"]

# Lista de precios
precios = [54.99, 17.81, 21.78, 45.89]

# Lista de existencias
existencias = [2450, 273, 1501, 998]

# Abrimos el archivo para escribir el reporte
f = open("reporte_productos.txt", "w")

# Unimos las tres listas como una sola (de tuplas)
productos = zip(nombres, precios, existencias)

# Calculamos el separador
f.write("+{}+{}+{}+".format("-" * 22, "-" * 10, "-" * 13))
f.write("\n")
f.write("| {:20} | {:8} | {:10} |".format("NOMBRE", "PRECIO", "EXISTENCIAS"))
f.write("\n")
f.write("+{}+{}+{}+".format("-" * 22, "-" * 10, "-" * 13))
f.write("\n")

for nombre, precio, existencias in productos:
  f.write("| {:20} | ${:>7.2f} | {:>11} |".format(nombre, precio, existencias))
  f.write("\n")

f.write("+{}+{}+{}+".format("-" * 22, "-" * 10, "-" * 13))

f.close()
````
