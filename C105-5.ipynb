## 1. Aquisición de los enlaces de los Departamentos

```python
import scrapydo

scrapydo.setup()
```

```python
def parser(response):
  print(response.url)

  # 1. Buscar los enlaces a los departamentos
  # --- <ul class="departmentMenu">
  # --- <li><a href="URL">NOMBRE DEPARTAMENTO</a></li>

  # @class='departmentMenu'
  # starts-with(@class, 'departmentMenu')
  # contains(@class, 'departmentMenu')

  for link in response.xpath("//ul[@class='departmentMenu']/li/a"):
    url = link.xpath("./@href").get()
    nombre = link.xpath("./text()").get()

    yield { "nombre": nombre, "url": f"https://www.homedepot.com.mx{url}" }

departamentos = scrapydo.crawl("https://www.homedepot.com.mx/", callback=parser)

import pandas as pd

departamentos = pd.DataFrame(departamentos)

departamentos.to_csv("homedepot_departamentos.csv", index=False)

departamentos.sample(5)
```
